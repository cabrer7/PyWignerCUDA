# RECIPROCAL - PHASE SPACE INVERTED OSCILLATOR

```python
from scipy.special import laguerre
import numpy as np
import scipy.fftpack as fftpack
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import time
import h5py
```

### Potential

```python
mass = 1.
omega = 1

def Potential(x):
    """
    Potential  function   
    """
    return -0.5*mass*omega**2*x**2
            
def dPotentialdx(x):
    """
    dPotential/dx  function   
    """
    return -mass*omega**2*x
```

### Parameters in atomic units

```python

X_gridDIM =  512     # Discretization grid size in X
P_gridDIM =  512    # Discretization grid size in P
        
X_amplitude  = 20    # Window range -X_amplitude to X_amplitude
P_amplitude  = 20   # Window range -P_amplitude to P_amplitude
        
dt= 0.005             # Time discretization
timeSteps = 200   # Number of iterations
hbar = 1    
timeRange     =  np.array(range(0,timeSteps+1))*dt 
        
print ' Simulation time = ', np.max(timeRange) ,'(a.u.)'
```

### Resolution & coordinates

```python
# Discretization resolution

dX =  2.*X_amplitude/float(X_gridDIM)
dP =  2.*P_amplitude/float(P_gridDIM)

dTheta  = 2.*np.pi/(2.*P_amplitude)
Theta_amplitude = dTheta*P_gridDIM/2.
dLambda = 2.*np.pi/(2.*X_amplitude)
Lambda_amplitude = dLambda*X_gridDIM/2.

# vectors with range of coordinates

X_range      =  np.linspace(-X_amplitude      , X_amplitude  -dX , X_gridDIM )
Lambda_range =  np.linspace(-Lambda_amplitude , Lambda_amplitude-dLambda  , X_gridDIM)

Theta_range  = np.linspace(-Theta_amplitude  , Theta_amplitude - dTheta , P_gridDIM)
P_range      = np.linspace(-P_amplitude      , P_amplitude-dP           , P_gridDIM)	

# matrices of grid of coordinates [p,x]

X      = fftpack.fftshift(X_range)[np.newaxis,:]
Theta  = fftpack.fftshift(Theta_range)[:,np.newaxis]

Lambda = fftpack.fftshift(Lambda_range)[np.newaxis,:]
P      = fftpack.fftshift(P_range)[:,np.newaxis]

print "Resolution features : "
print ""
print "dX =", dX ,"                            dP = ",dP
print ""
print "+-X_amplitude =",X_amplitude, "                            +-P_amplitude =",P_amplitude
print ""
print "dLambda =" ,dLambda ,"                   dTheta =",dTheta
print ""
print "+-Lambda_amplitude =",Lambda_amplitude,"         +-Theta_amplitude =",Theta_amplitude
```

### Plotting potential

```python
plt.plot( X_range, Potential(X_range) ,label = '$V \\ =\\ - \\ \\frac{1}{2} m \omega^2 x^2 $' )
xlabel('$x$')
legend(loc = "upper right")
grid('on')
```

### Initial Wigner function

```python
def Wigner_gaussian(x,p,s):
    """
    Wigner function of a Gaussian
    x,p  : center of wave packet
    s    : Standard deviation 
    """
    return np.exp(  -( (X-x)/s )**2 -( s*(P-p) )**2  ) + 0j
```

```python
#  Center of the wave packet

x_init = 0
p_init = 0

# Standard deviation in x

s = 1.

# Normalization

W_init = Wigner_gaussian( x_init, p_init ,s )
norm   = np.sum( W_init )*dX*dP
W_init /= norm

# Initial Ambiguity Function

A_init = fftpack.fft2( W_init , axes=(0, 1), overwrite_x= True)


# Features

print "Eigen_energy",(np.sum(P**2*W_init)*dP*dX)/(2*mass) + np.sum(Potential(X)*W_init)*dP*dX
print "W, Renormalization norm :",np.sum( W_init )*dX*dP
print "Purity :", 2*np.pi*np.sum(W_init**2)*dX*dP
```

```python
#--- Features of the initial ambiguity function

Real_volume_init = np.sum(np.real(A_init))*dLambda*dTheta
Imag_volume_init = np.sum(np.imag(A_init))*dLambda*dTheta
Pos_volume_init = np.sum(A_init[np.nonzero( A_init.real > 0. )])*dLambda*dTheta
Neg_volume_init = np.sum(A_init[np.nonzero( A_init.real < 0. )])*dLambda*dTheta

print "A, real volume:",Real_volume_init ,"   A, Imag volume:",Imag_volume_init
print "A, Pos volume:" ,Pos_volume_init  ,"   A, Neg volume:",Neg_volume_init
```

### Marginals

```python
# Marginals x,p space
W_x = np.sum(W_init,axis=0)*dP
W_p = np.sum(W_init,axis=1)*dX
plot( X_range ,fftpack.fftshift(W_x).real  )
plot( P_range ,fftpack.fftshift(W_p).real  )
```

```python
# Marginals "Lambda,Theta" space

A_lambda = np.sum(A_init,axis=0)*dTheta
A_Theta  = np.sum(A_init,axis=1)*dLambda
plot( Lambda_range ,fftpack.fftshift(A_lambda).real , label = '$ A_{\lambda}$' )
plot( Theta_range ,fftpack.fftshift(A_Theta).real , label = '$ A_{\theta}$' )
xlim([-10,10])

legend(loc = "upper right")
```

### Analytical reciprocal Newton's equation of motion

```python
def R_Newton_dots(time , n_newton):
    
    """This functions compute the position and momentum of "n_newton**2"
       particles at specific time
       
       time     :  time for wich will be compute lambda,theta for
       (n_newton**2) particles 
       
       n_newton : (int) the square of this amount is the 
       number of particles """
    
    lambda_init = 0
    theta_init  = 0
    
    Lambda_Range_Newton = np.linspace( lambda_init -3.5 , lambda_init + 3.5 , n_newton )   
    Theta_Range_Newton = np.linspace( theta_init -3.5 , theta_init + 3.5 , n_newton )
    
    data_Lambda_Newton = []
    data_Theta_Newton = []
    
    for Index_Lambda in range (0,n_newton):
        for Index_Theta in range(0,n_newton):
                        
            Lambda_Newton = Lambda_Range_Newton[Index_Lambda]*np.cosh(time) - Theta_Range_Newton[Index_Theta]*np.sinh(time)
            Theta_Newton  = -Lambda_Range_Newton[Index_Lambda]*np.sinh(time) + Theta_Range_Newton[Index_Theta]*np.cosh(time)
            
            data_Lambda_Newton.append(Lambda_Newton)
            data_Theta_Newton.append(Theta_Newton)
            
    return np.array(data_Lambda_Newton),np.array(data_Theta_Newton)
```

### Wigner Propagator 

```python
#------Propagators factors

expPLambda = np.exp(-1j*dt*P*Lambda/mass)
expPotential = np.exp( -1j*dt*(Potential(X-hbar*Theta/2.) - Potential(X+hbar*Theta/2.))/hbar )
```

```python
#--- List to save data for Ehrenfest Theorems ,Energy and Uncertainty principle:

data_Expectation_X=[np.sum(X*W_init)*dP*dX]
data_Expectation_X2=[np.sum(X**2*W_init)*dP*dX]
data_Expectation_P=[np.sum(P*W_init)*dP*dX]
data_Expectation_P2=[np.sum(P**2*W_init)*dP*dX]
data_Expectation_XP=[np.sum(X*P*W_init)*dP*dX]
data_Expectation_F=[-1*np.sum( dPotentialdx(X)*W_init)*dP*dX]
data_Expectation_XdVdX=[np.sum(X*dPotentialdx(X)*W_init)*dP*dX]
data_Expectation_PdVdX=[np.sum(P*dPotentialdx(X)*W_init)*dP*dX]
data_Expectation_V=[np.sum(Potential(X)*W_init)*dP*dX]
data_NegArea = [np.sum(W_init[np.nonzero( W_init.real < 0. )]).real*dP*dX]
data_Integral_W2 = [np.sum(W_init**2)*dX*dP]

#--- Propagator
        
W = W_init.copy() + 0j

for tIndex in range(1,timeSteps+1):
    if tIndex%1000==0:
        print '  tIndex  = ', tIndex
            
    t = (tIndex)*dt
    
    # p x  ->  p lambda 
    
    W = fftpack.fft( W, axis = 1 )
    W *=expPLambda

    # p lambda  ->  p x
    W = fftpack.ifft( W, axis = 1 )
    
    # p x -> theta x
    W = fftpack.fft( W, axis = 0 )
    W *= expPotential

    # theta x  ->  p x
    W = fftpack.ifft( W, axis = 0 )
    
    # p x  -> theta lambda
    
    #A = fftpack.fft2( W , axes=(0, 1), overwrite_x= True)
    
    #--- Compute expectations values
        
    Expectation_X = np.sum(X*W)*dP*dX
    Expectation_X2=np.sum(X**2*W)*dP*dX
    Expectation_P = np.sum(P*W)*dP*dX
    Expectation_P2= np.sum(P**2*W)*dP*dX
    Expectation_XP= np.sum(X*P*W)*dP*dX
    Expectation_F= -1*np.sum( dPotentialdx(X)*W)*dP*dX 
    Expectation_XdVdX=np.sum(X*dPotentialdx(X)*W)*dP*dX
    Expectation_V=np.sum(Potential(X)*W)*dP*dX
    NegArea = np.sum(W[np.nonzero( W.real < 0. )]).real*dP*dX
        
    #--- Make data list
        
    data_Expectation_X.append(Expectation_X)  # 1
    data_Expectation_X2.append(Expectation_X2) #2
    data_Expectation_P.append(Expectation_P)   #3
    data_Expectation_P2.append(Expectation_P2) #4A
    data_Expectation_XP.append(Expectation_XP) #5
    data_Expectation_F.append(Expectation_F)   #6
    data_Expectation_XdVdX.append(Expectation_XdVdX) #7
    data_Expectation_V.append(Expectation_V) #8
    data_NegArea.append(NegArea) #9
```

```python
W_end = W + 0j
print " W_end, Renormalization norm: ",np.sum(W_end)*dX*dP

# Final ambiguity function A

A_end = fftpack.fft2( W_end , axes=(0, 1), overwrite_x= True)
```

```python
#---Features of final  ambiguity function A_end

Real_volume_end = np.sum(np.real(A_end))*dLambda*dTheta
Imag_volume_end = np.sum(np.imag(A_end))*dLambda*dTheta
Pos_volume_end = np.sum(A_init[np.nonzero( A_end.real > 0. )])*dLambda*dTheta
Neg_volume_end = np.sum(A_init[np.nonzero( A_end.real < 0. )])*dLambda*dTheta

print "A, real volume:",Real_volume_end ,"   A, Imag volume:",Imag_volume_end
print "A, Pos volume:" ,Pos_volume_end  ,"   A, Neg volume:",Neg_volume_end
```
