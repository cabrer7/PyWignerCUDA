```python
# Defining the instance 

class Rho2D_Dissipation_X2(Rho2D_Dissipation):
    """
    Instance for dissipative dynamics     
    """
    def __init__(self):
        
        self.omega = 0.2
        self.x_init = -0.
        self.p_init =  2.5
        n = 0 
        mass = 1
        dampingType = 'Ohmic'
        
        constantX2 = 0.5*mass*self.omega**2
        # Potential and derivative of potential        
        self.potentialString = '300*tanh( ({0}*pow(x,2) + 0.01*pow(x,4))/300. )'.format( constantX2 )

        self.dPotentialString = 'x*( 0.04*x*x + 2*{0} )/power( cosh( ({0}*pow(x,2) + 0.01*pow(x,4))/300. ),2 ) '.format( constantX2 )
        
        # Data output file
        self.fileName = '/home/rcabrera/DATA/Rho2D/'+dampingType+'/L{0}_epsilon{1}_grid{2}_amplitude{3}'.format(lambdaBar,epsilon,X_gridDIM,X_amplitude)
        self.fileName += '/Rho2D_ODMDamping_X2-X4_PInit{0}_Omega{1}.hdf5'.format(self.p_init,self.omega)
        
        # Lindbladian operator A
        self.A_fileName = '/home/rcabrera/Documents/source/python/PyWignerCUDA/instances/Rho2D/'+dampingType
        self.A_fileName += '/L{0}_epsilon{1}_grid{2}_amplitude{3}'.format(lambdaBar,epsilon,X_gridDIM,X_amplitude)
        self.A_fileName += '/Lindblad_A.hdf5'
                
        Rho2D_Dissipation.__init__(
                self,X_gridDIM,X_amplitude,dt,timeSteps,skipFrames,mass,gamma,D_Theta) 
        
        # Setting initial state   
        self.Rho_init = self.Rho_HarmonicOscillator(n, self.omega*5 ,self.x_init, self.p_init)
                                                                  
```

```python
def EnergyWigner(n):
    Rho = instance.LoadRho(instance.fileName, n ) 
    W   = instance.DensityMatrix_To_Wigner(Rho)

    X_Range = instance.X_Range()/np.sqrt(2)
    P_Range = instance.P_Range()/np.sqrt(2)

    X = X_Range[np.newaxis,:]
    P = P_Range[:,np.newaxis]
    
    H = P**2/(2*instance.mass) + instance.Potential(0,X)
    
    return np.sum( W*H )*instance.dX*instance.dP/2.
```

```python
skipFactor =  10

energyW_list = []
for n in range( 0, instance.timeSteps, skipFactor*instance.skipFrames ):
    print ' n = ',n
    energyW_list.append( EnergyWigner(n)  )
    
energyW_list  = np.array( energyW_list )

energyW_timeLine  = instance.dt*np.array( range( 0, instance.timeSteps, skipFactor*instance.skipFrames ) )
```

```python
def PlotWigner( W2 , messages=True,saveFigure=False):
    global_min = -0.1
    global_max =  0.1
    
    zero_position =  abs( global_min) / (abs( global_max) + abs(global_min)) 
    wigner_cdict = {'red' 	: 	((0., 0., 0.),
                                (zero_position, 1., 1.), 
                                (1., 1., 1.)),
                        'green' :	((0., 0., 0.),
                                (zero_position, 1., 1.),
                                (1., 0., 0.)),
                        'blue'	:	((0., 1., 1.),
                                (zero_position, 1., 1.),
                                (1., 0., 0.)) }
    
    wigner_cmap = matplotlib.colors.LinearSegmentedColormap('wigner_colormap', wigner_cdict, 512)
    #wigner_cmap = LinearSegmentedColormap('wigner_colormap', wigner_cdict, 512)
    
    plt.clf()
    fig, ax = plt.subplots(figsize=(10, 6))
    
    dp = instance.dP
    dx = instance.dX
    
    p_min = -instance.P_amplitude + dp/2.
    p_max =  instance.P_amplitude - dp/2.   
        
    x_min = -instance.X_amplitude + dx/2.
    x_max =  instance.X_amplitude - dx/2.
    
    
    norm = np.sum(W2)*dp*dx
    W2 /= norm
        
    cax = ax.imshow(  W2.real  ,
           extent = [ x_min , x_max , p_min , p_max ], 
           vmin= global_min, vmax=global_max,
           origin='lower',interpolation='none',cmap=wigner_cmap)
    
    X = instance.X_Range()[np.newaxis,:]
    P = instance.P_Range()[:,np.newaxis]
    
    H = P**2/(2*instance.mass) + instance.Potential(0,X)
       
    ax.contour( H ,  energySpectrum , origin='lower',extent=[x_min,x_max,p_min,p_max],
               linewidths=0.25,colors='k')
    
    """H = P**2/(2*instance.mass) + 0.5*instance.mass*X**2
    ax.contour( H ,  np.arange( 0.5, 10, 1 ) , origin='lower',extent=[x_min,x_max,p_min,p_max],
               linewidths=0.25,colors='k')"""
    
    ax.set_xlim( -6,6 )
    ax.set_ylim( -6,6 )
    ax.set_xlabel('x')
    ax.set_ylabel('p')
    
    ax.grid('on')
    
    ax.set_aspect(1)
    
    cbar = fig.colorbar( cax , ticks=[-0.3, -0.2,-0.1, 0, 0.1, 0.2 , 0.3])

    if saveFigure != False:
        fig.savefig( saveFigure  )
        
        
    if messages == True:
        print ' max W = ', np.max(W2.real)
        print ' min W = ', np.min(W2.real)
        print ' norm =  ', np.sum(W2).real*dp*dx
        print ' Energy =', np.sum( H*W2  )*dx*dp
        
        
    return fig
```

```python
def PlotWignerScaled( W2 , messages=True,saveFigure=False):
    global_min = -0.08
    global_max =  0.08
        
    zero_position =  abs( global_min) / (abs( global_max) + abs(global_min)) 
    wigner_cdict = {'red' 	: 	((0., 0., 0.),
                                (zero_position, 1., 1.), 
                                (1., 1., 1.)),
                        'green' :	((0., 0., 0.),
                                (zero_position, 1., 1.),
                                (1., 0., 0.)),
                        'blue'	:	((0., 1., 1.),
                                (zero_position, 1., 1.),
                                (1., 0., 0.)) }
    
    wigner_cmap = matplotlib.colors.LinearSegmentedColormap('wigner_colormap', wigner_cdict, 512)
    
    plt.clf()
    fig, ax = plt.subplots( figsize=(10, 8) )
    
    X_Range = instance.X_Range()/np.sqrt(2)
    P_Range = instance.P_Range()/np.sqrt(2)
    
    X = X_Range[np.newaxis,:]
    P = P_Range[:,np.newaxis]
    
    p_min = P.min();   p_max = P.max()
    x_min = X.min();   x_max = X.max()
    
    dp = P_Range[1] - P_Range[0]
    dx = X_Range[1] - X_Range[0]
    
    norm = np.sum(W2)*dp*dx
    W2 /= norm
        
    cax = ax.imshow(  W2.real  ,
           extent = [ x_min , x_max , p_min , p_max ], 
           vmin= global_min, vmax=global_max,
           origin='lower',interpolation='none',cmap=wigner_cmap)
    
    H = P**2/(2*instance.mass) + instance.Potential(0,X)
       
    CS = ax.contour( H ,  energySpectrum[0:12] , origin='lower',extent=[x_min,x_max,p_min,p_max],
               linewidths=0.25,colors='green')
    
        
    ax.clabel( CS, inline=10, fontsize=10, fmt='%1.2f')
    
    ax.set_xlim( -5,5 )
    ax.set_ylim( -5,5 )
    ax.set_xlabel('x')
    ax.set_ylabel('p')
    
    ax.grid('on')
    
    ax.set_aspect(1)
    
    cbar = fig.colorbar( cax , ticks=[-0.3, -0.2,-0.1, 0, 0.1, 0.2 , 0.3])

    if saveFigure != False:
        fig.savefig( saveFigure  )
        
        
    if messages == True:
        print ' max W = ', np.max(W2.real)
        print ' min W = ', np.min(W2.real)
        print ' norm =  ', np.sum(W2).real*dp*dx
        print ' Energy (Wigner) =', np.sum( H*W2  ).real*dx*dp
        
```

## Projections

```python
w,v = linalg.eigh( instance.H )
```

```python
nMax=32

projectionsOdd = []

for n in range(1,nMax,2):
    Rho   = instance.LoadRho(instance.fileName, instance.timeSteps)
    projectionsOdd.append( np.trace( Rho.dot( instance.Psi_To_Rho(v[:,n]) ) )  )
    
projectionsOdd = np.array(projectionsOdd)
```

```python
projectionsEven = []

for n in range(0,nMax,2):
    RhoHO = instance.Psi_To_Rho(v[:,n])
    Rho   = instance.LoadRho(instance.fileName, instance.timeSteps)
    projectionsEven.append( np.trace( Rho.dot(RhoHO) )  )
    
projectionsEven = np.array(projectionsEven)
```

```python
plt.semilogy(  range(1,nMax,2), projectionsOdd.real   ,'.', label='Odd states')
plt.semilogy(  range(0,nMax,2), projectionsEven.real   ,'r.',label='Even states')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.xlabel('n')
```

```python
RhoSum = np.zeros( (instance.gridDIM,instance.gridDIM) , dtype=np.complex128)

for n in range(0,nMax/2):
    RhoSum += projectionsEven[n]*instance.Psi_To_Rho(v[:,2*n])
    

for n in range(1,nMax/2):
    RhoSum += projectionsOdd[n]*instance.Psi_To_Rho(v[:,2*n+1])  
    
WSum = instance.DensityMatrix_To_Wigner(  RhoSum )
```

```python
energySum = 0
for n in range(0,nMax/2):
    energySum += projectionsEven[n]*energySpectrum[2*n]
    

for n in range(1,nMax/2):
    energySum += projectionsOdd[n]*energySpectrum[2*n+1]
    
print ' Energy Sum = ', energySum.real    
```
